```python
import pandas as pd
# import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("spotify_data.csv")

data.info()
```

```python
#Checking for duplicate rows
data[data.duplicated()]

#Checking for negative streams
(data["Streams"] < 0).any()

#Changing the data type of Date field 
data["Date"] = pd.to_datetime(data["Date"])
```

### Top 20 Most Streamed Artist from the Daily World Rankings of Spotify Philippines in 2017

```python
df = data[data["Region"] == 'ph']
df[df["Date"].dt.year == 2017].groupby(["Artist"], as_index=False).agg({"Streams": "sum"}).sort_values("Streams", ascending=False).head(20)
```

### Top 20 Songs that Appeared the Most in the Daily World Rankings of Spotify Philippines

```python
data[data["Region"] == "ph"]["Track Name"].value_counts().sort_values(ascending=False).head(20)
```

### Top 20 Artist with the Most Number of Songs in Daily Rankings of Spotify Philippines

```python
data[data["Region"] == "ph"][["Artist", "Track Name"]].groupby(["Artist"]).count().sort_values("Track Name", ascending=False).head(20).rename(columns={"Track Name": "Count"})
```

### Top 20 Artist with Most Number of Unique Songs in Daily Rankings of Spotify Philippines

```python
df = data[data["Region"] == "ph"][["Artist", "Track Name"]].groupby("Artist").nunique().sort_values("Track Name", ascending=False).head(20)
df.rename(columns={"Track Name": "Unique_Songs_Count"}, inplace=True)
df
```

### Top 20 Most Streamed Songs in the Philippines 2017

```python
df2 = data[data["Date"].dt.year == 2017]
df3 = df2[df2["Region"] == 'ph'].groupby(["Artist", "Track Name"], as_index=False).agg({"Streams": "sum"}).sort_values(["Streams"], ascending=False).head(20)
df3
```
